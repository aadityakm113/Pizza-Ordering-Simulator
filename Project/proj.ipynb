```python
'''
Pizza Ordering Simulator 
By Aaditya Kumar Muktavarapu
HU21CSEN0100580
amuktava@gitam.in
'''
from msilib.schema import CheckBox, Font
from pkgutil import extend_path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tkinter
from tkinter import *
import sqlite3

#Database Creation
conn=sqlite3.connect("Database.db")

#Cursor Creation
c=conn.cursor()

#Table Creation

#Pricing Table
c.execute("""
            CREATE TABLE IF NOT EXISTS Pricing("S.No" int,
            "Base/Topping" text,"Price" text)
            """)

#Price dictionary
price = {'Sausage': "₹40", 'Pepperoni': "₹40", 'Chicken': "₹40", 'Olives': "₹35",'Extra Cheese':'₹30','Jalepenos':"₹35"}
#Entering Values in Pricing Table
pricevalues=[
    ("1","Small","₹99"),
    ("2","Medium","₹199"),
    ("3","Large","₹399"),
    ("4","Olives","₹35"),
    ("5","Chicken","₹40"),
    ("6","Jalepenos","₹35"),
    ("7","Pepperoni","₹40"),
    ("8","Sausage","₹40"),
    ("9","Extra Cheese","30")
]
#c.executemany("INSERT into Pricing VALUES(?,?,?)",pricevalues)

#User Table
c.execute("""
            CREATE TABLE IF NOT EXISTS User("Phone number" int PRIMARY KEY,
            "Name" text,"Username" text, "Password" text)
            """)

conn.commit()

#home screen
home=Tk()
home.title("Aadi's Pizzeria")
home.geometry("500x300")
FONT=("Comic Sans MS",20,"bold")
lbl = Label(home, text = "Welcome to Aadi's Pizzeria",font=FONT)
lbl.pack()

#User Entries
l5=Label(home,text="Enter Username: ")
l5.pack()
si_username=Entry(home)
si_username.pack()
l6=Label(home,text="Enter Password: ")
l6.pack()
si_password=Entry(home)
si_password.pack()

#Order Window
def orderwin():
    for widget in home.winfo_children():
            widget.destroy()
    home.geometry("700x400")
    home.title("Order Your Pizza")

    orderfont=("",20,"bold")
    l7=Label(home,text="Order Your Pizza",anchor="center",font=orderfont)
    l7.pack()

    sizefont=("",14,"")
    l8=Label(home,text="Size",anchor="center",font=sizefont)
    l8.pack()

    #Size of the pizza base
    r=StringVar()
    #r.set()
    a=Radiobutton(home,text="Small (₹99)",variable=r,value=99,anchor="w")
    a.deselect()
    a.pack()
    b=Radiobutton(home,text="Medium (₹199)",variable=r,value=199,anchor="center")
    b.deselect()
    b.pack()
    c=Radiobutton(home,text="Large (₹399)",variable=r,value=399,anchor="e")
    c.deselect()
    c.pack()

    #Choice of toppings
    toppingfont=("",14,"")
    l9=Label(home,text="Toppings",anchor="center",font=toppingfont)
    l9.pack()

    #Non Veg
    sau=StringVar()
    s=Checkbutton(home,text="Sausage (₹40)",variable=sau,onvalue=40)
    s.deselect()
    s.pack()
    chi=StringVar()
    chick=Checkbutton(home,text="Chicken (₹40)",variable=chi,onvalue=40)
    chick.deselect()
    chick.pack()
    pe=StringVar()
    pep=Checkbutton(home,text="Pepperoni (₹40)",variable=pe,onvalue=40)
    pep.deselect()
    pep.pack()

    #Veg
    exch=StringVar()
    ec=Checkbutton(home,text="Extra Cheese (₹35)",variable=exch,onvalue=35)
    ec.deselect()
    ec.pack()
    ol=StringVar()
    o=Checkbutton(home,text="Olives (₹35)",variable=ol,onvalue=35)
    o.deselect()
    o.pack()
    jal=StringVar()
    j=Checkbutton(home,text="Jalepenos (₹35)",variable=jal,onvalue=35)
    j.deselect()
    j.pack()

    def place():
        show=Label(home,text=ol.get())
        show.pack()
    placeorder=Button(home,text="Place Order",command=place)
    placeorder.pack()



#Sign up Window
def suw():
    global signupwin
    signupwin=Toplevel()
    signupwin.geometry("500x300")
    signupwin.title("Create Account")
    
    #Text box to enter user details
    l=Label(signupwin,text="Enter Name: ")
    l.pack()
    name=Entry(signupwin)
    name.pack()
    l1=Label(signupwin,text="Enter Username: ")
    l1.pack()
    su_username=Entry(signupwin)
    su_username.pack()
    l2=Label(signupwin,text="Enter Password: ")
    l2.pack()
    su_password=Entry(signupwin)
    su_password.pack()
    l3=Label(signupwin,text="Re-enter your password: ")
    l3.pack()
    repass=Entry(signupwin)
    repass.pack()
    l4=Label(signupwin,text="Enter your Phone Number: ")
    l4.pack()
    ph=Entry(signupwin)
    ph.pack()
    

    #saving Details to database
    def save():
        
        getname=name.get()
        getus=su_username.get()
        getpass=su_password.get()
        getrepass=repass.get()
        getph=ph.get()
        
        values=(getph,getname,getus,getpass)
        if getpass==getrepass:
            c.execute("INSERT into User VALUES(?,?,?,?)",values)
            conn.commit()
            orderwin()
                
        else:
            elbl=Label(signupwin,text="Passwords not Matching")
            elbl.pack()

        #ul=Label(signupwin,text=getus)
        #ul.pack()
    store=Button(signupwin,text="Save Details",command=save)
    store.pack()

#retrieving username and password from database and comparing with user inputted credentials
def check():
    getluser=si_username.get()
    getlpass=si_password.get()
    c.execute("SELECT * from User WHERE Username=(?)",(getluser,))
    data=c.fetchall()
    conn.commit()
    if getlpass==data[0][3]:
        #suclbl=Label(home,text="Success!")
        #suclbl.pack()
        orderwin()
    else:
        elbl1=Label(home,text="Invalid Credentials")
        elbl1.pack()

SI=Button(home,text="Sign in",command=check)
SI.pack()
SU=Button(home,text="Sign up",command=suw)
SU.pack()

home.mainloop()
```
